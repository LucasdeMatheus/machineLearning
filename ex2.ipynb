```python
# ----------------------- bibliotecas ----------------------- 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# ---------------------------- Obtendo resultados ----------------------------

data = np.array([
    [10, 20],  # Dia 1
    [15, 30],  # Dia 2
    [50, 5],   # Dia 3
    [60, 10],  # Dia 4
    [8, 25]    # Dia 5
])

kmeans = KMeans(n_clusters=3, random_state=0)

kmeans.fit(data)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print(labels, centroids)
```

```python
# separando os dados

# dias

dias = len(data)

# Separando as colunas para Produto A e Produto B
produto_a = data[:, 0]
produto_b = data[:, 1]

# separando os centroides

centroid_1 = centroids[0]

centroid_2 = centroids[1]

centroid_3 = centroids[2] 

print("dias: ", dias, "Produtos A: ", produto_a,"Produtos B: ", produto_b, "Centroid 1: ", centroid_1, "Centroid 2: ", centroid_2, "Centroid 3: ", centroid_3,)
```

```python
# criando grafico
fig, ax = plt.subplots()

label = ['produto A', 'produto B']

bar_colors = ['tab:red', 'tab:blue']
for i in range(dias):
    ax.bar(i - 0.2, produto_a[i], width=0.4, label=label[0] if i == 0 else "", color=bar_colors[0])
    ax.bar(i - 0.2, produto_a[i], width=0.4, label=label[1] if i == 0 else "", color=bar_colors[1])

ax.set_xlabel('Dia')
ax.set_ylabel('Vendas')
ax.set_title('Vendas de Produto A e Produto B por Dia')
ax.set_xticks(dias)
ax.set_xticklabels(dias)

plt.show()
```

```python

# Ajuste para fazer as barras aparecerem lado a lado por dia
# Aqui, vamos usar o np.array() para ajustar as vendas em formato apropriado
for i in range(len(produto_a)):  # Ajuste para cada dia
    ax.bar(i - 0.2, produto_a[i], width=0.4, label='Produto A' if i == 0 else "", color='tab:red')
    ax.bar(i + 0.2, produto_b[i], width=0.4, label='Produto B' if i == 0 else "", color='tab:blue')

# Configurações do gráfico


# Exibindo a legenda e o gráfico
ax.legend()
plt.show()
```
